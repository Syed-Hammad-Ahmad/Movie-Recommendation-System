```python
#EXTARCT GENRES NAMES
#IMPORT REQUIRED LIBRARIES
import ast
#ast.literal_eval() is used to safely convert strings into python objects.
```

```python
#Now we create a function to extract the genres name from the genres column and
#apply that function to movies['genres']
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
movies['genres'] = movies['genres'].apply(convert)
```

```python
#CONVERT OVERVIEW COLUMN INTO LIST
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
# Now we want to remove spaces between names of every columns
# like "Sam Worthington" should be "SamWorthington"
# similarly "Science Fiction" should be "ScienceFiction"

# For genres column
movies['genre'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])

# For keywords column
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])

# For cast column
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])

# For crew column
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
# Now we will create a new column tags which will contain all the important information about a movie
movies['tags'] = (movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'])
```

```python
# Now we will create a new dataframe which will contain only movie_id, title and tags column 
# because we will use these columns for our recommendation system.
new_movies = movies [['movie_id','title', 'tags']]
```

```python
print(new_movies.head())
```

```python
# new_movies['tags'] = new_movies['tags'].apply(lambda x: " ".join(x)) WILL WORKS FIRST TIME 
# BUT SECOND TIME GIVES ERROR BECAUSE THE COLUMN IS NO LONGER LISTS; IT'S ALREADY STRINGS.
# IF YOU RUN IT AGAIN, THE TAGS COLUMN WILL THEN JOIN THE LETTERS OF THE STRING WITH A " " (gap)
# EACH TIME YOU RUN IT, IT WILL JOIN THE LETTERS OF THE STRING WITH A " " (gap) 
# AND EVENTUALLY YOU WILL GET A STRING WITH SPACES BETWEEN EACH LETTER.

# ------------------------------------------------------------------------------------------
# IF YOOU BY MISTAKE RUN IT AGAIN, THEN GO BACK TO CELL WHERE YOU CREATE NEW_MOVIES['TAGS']
# AND RUN THAT CELL AGAIN IN ORDER.
# ------------------------------------------------------------------------------------------

# Since the ML text processing model cannot understand list
# And we have to use the model CountVectorizer which takes string as input
# therefore we will convert the list of tags into string

new_movies['tags'] = new_movies['tags'].apply(lambda x: " ".join(x))

# Now we have to lower the text in tags column 
# because the model cannot understand uppercase and lowercase letters as same.

new_movies['tags'] = new_movies['tags'].apply(lambda x: x.lower())
```

```python
# ==========================================
# TEXT VECTORIZATION
# ==========================================
```

```python
# Computers cannot compare text directly.
# We need to convert each movie's tags into a vector (a list of numbers).

from sklearn.feature_extraction.text import CountVectorizer 

# Step 1 : Create CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')

# What does this mean? max_features=5000
# Keep only the 5000 most important words.
# Without this, we might get 20,000+ unique words

# stop_words='english' -> Remove common words like: the, is, are, was, in, on, or, etc
# since these words don't help recommendations/similarity
# therfore it is better to remove them
```

```python
# Step 2 : Transform Tags into Vectors
vector = cv.fit_transform(new_movies['tags']).toarray()
# vectors[i] contains the numerical representation of the i-th movie,
# which can then be used for similarity calculations.
print(vector.shape)
# this means 4806 movies, 5000 features(words)
```

```python
# To see which words were chosen
cv.get_feature_names_out()
```

```python
# =========================================
# -------------MOST IMPORTANT--------------
# <<<<<<<<<<<COSINE SIMILARITY>>>>>>>>>>>>>
# =========================================
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
# This show that every movie is compared with every movie.
print(similarity.shape)
print(vector.shape)
```

```python
#----------PROBLEM-----------
#in order to search for any movie's recommendation
#we must know its exact name in exact manner.
# eg. "Spider-Man: Homecoming" is correct but "spiderman homecoming" is wrong
# eg. "Avatar" is correct but "avater" is wrong
#----------SOLUTION----------
# In order to solve this problem, we have 2 methods:-
# 1. Normalize Titles : We will create a fn that converts title to lowercase, removes punctuations,
# removes ,'-:   , removes extra spaces.
# 2. Fuzzy Matching : Suppose a user writes "dark knight" instead of "The Dark Knight" or 
# writes "avngrs" instead of "The Avengers"
# We can find the closest matching title automatically using get_close_matches.

import re
from difflib import get_close_matches

def clean_title(title):
    title = title.lower()
    title = re.sub(r'[^a-z0-9\s]','',title)
    title = " ".join(title.split())
    return title
```

```python
new_movies['clean_title']=new_movies['title'].apply(clean_title)
```

```python
#print(new_movies[['title','clean_title']].head())
```

```python
def recommend(movie):
    movie = clean_title(movie)
    if movie in new_movies['clean_title'].values:
        movie_index = new_movies[new_movies['clean_title']==movie].index[0]
        original_title = new_movies.iloc[movie_index].title
        print(f"Showing results for : {original_title}")
    
    else:
        closest_match = get_close_matches(movie, new_movies['clean_title'], n=1, cutoff=0.75)
        if not closest_match:     # means if closest_match == []
            print("Movie not found")
            return
        matched_movie = closest_match[0]
        movie_index = new_movies[new_movies['clean_title'] == matched_movie].index[0]
        original_title = new_movies.iloc[movie_index].title
        print(f"Showing results for : {original_title}")

    distances = similarity[movie_index]

    movie_list = sorted(list(enumerate(distances)), reverse=True, key = lambda x:x[1])

    # Suppose after sorting, movie_list looks like:
    # movie_list = [(154, 1.00),(23, 0.89),(45, 0.85),(67, 0.81),(12, 0.79),(90, 0.77)]
    # where each tuple = (movie_index, similarity_score)
    # so movie_list[1:6] returns [(23, 0.89),(45, 0.85),(67, 0.81),(12, 0.79),(90, 0.77)]
    # But we also want numbering, so we use enumerate(movie_list[1:6]) which produces:-
    # (0, (23, 0.89)) (1, (45, 0.85)) (2, (67, 0.81)) (3, (12, 0.79)) (4, (90, 0.77))
    # with start=1, output becomes:-
    # (1, (23, 0.89)) (2, (45, 0.85)) (3, (67, 0.81)) (4, (12, 0.79)) (5, (90, 0.77))
    # Loop variables :- for idx, i in enumerate(movie_list[1:6], start=1):
    # first iteration : idx = 1, i = (23, 0.89)
    # new_movies.iloc[i[0]] -> new_movies.iloc[23] -> gets row number 23 from the dataframe.
    # second iteration : idx = 2, i = (45, 0.85)
    # new_movies.iloc[i[0]] -> new_movies.iloc[45] -> gets row number 45 from the dataframe.
    # third iteration : idx = 3, i = (67, 0.81)
    # new_movies.iloc[i[0]] -> new_movies.iloc[67] -> gets row number 67 from the dataframe.
    # fourth iteration : idx = 4, i = (12, 0.79)
    # new_movies.iloc[i[0]] -> new_movies.iloc[12] -> gets row number 12 from the dataframe.
    # .title gives title of row i
    for idx, i in enumerate(movie_list[1:6], start=1):
        print(f"{idx}. {new_movies.iloc[i[0]].title}")
```

```python
recommend("avengers age of ultron")
```
